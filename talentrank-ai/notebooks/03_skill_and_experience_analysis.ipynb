# TalentRank AI: Skill & Experience Analysis

Analyzing the skills array and its relation to career history.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '../')
from src.utils.io import load_jsonl
from src.config.settings import get_settings

candidates = list(load_jsonl(get_settings().paths.candidates_jsonl, max_records=10000))
```

```
all_skills = []
for c in candidates:
    all_skills.extend([s.get('name', '').lower() for s in c.get('skills', [])])

skill_counts = pd.Series(all_skills).value_counts().head(20)
plt.figure(figsize=(12, 8))
sns.barplot(y=skill_counts.index, x=skill_counts.values)
plt.title('Top 20 Skills Overall')
plt.show()
```
