# TalentRank AI: Dataset Overview

This notebook provides a high-level overview of the candidate dataset (100k records). We will analyze the missingness, schema compliance, and basic distributions of the core fields.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path

import sys
sys.path.insert(0, '../')
from src.config.settings import get_settings
from src.utils.io import load_jsonl

settings = get_settings()
sns.set_theme(style='whitegrid')
```

## 1. Load Data Sample
We load a sample of the data to avoid memory issues during initial exploration.

```
candidates = list(load_jsonl(settings.paths.candidates_jsonl, max_records=10000))
print(f'Loaded {len(candidates)} candidates.')
```

## 2. Basic Profile Distributions

```
# Experience Distribution
experience = [c['profile'].get('years_of_experience', 0) for c in candidates]
plt.figure(figsize=(10, 6))
sns.histplot(experience, bins=30, kde=True)
plt.title('Distribution of Years of Experience')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```
# Current Industry Distribution
industries = [c['profile'].get('current_industry', 'Unknown') for c in candidates]
ind_series = pd.Series(industries).value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=ind_series.index, x=ind_series.values)
plt.title('Top 10 Current Industries')
plt.show()
```
