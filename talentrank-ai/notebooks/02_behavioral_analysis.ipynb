# TalentRank AI: Behavioral Signal Analysis

Analyzing the `redrob_signals` object. These signals are critical for determining if a candidate is hirable (active, responsive, open to work).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

sys.path.insert(0, '../')
from src.config.settings import get_settings
from src.utils.io import load_jsonl

settings = get_settings()
sns.set_theme(style='whitegrid')
```

```
candidates = list(load_jsonl(settings.paths.candidates_jsonl, max_records=10000))
signals = [c.get('redrob_signals', {}) for c in candidates]
df = pd.DataFrame(signals)
df.head()
```

## Response Rates & Activity

```
plt.figure(figsize=(10, 6))
sns.histplot(df['recruiter_response_rate'].dropna(), bins=20)
plt.title('Recruiter Response Rate Distribution')
plt.xlabel('Response Rate (0-1)')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.histplot(df['profile_completeness_score'].dropna(), bins=20)
plt.title('Profile Completeness Score')
plt.show()
```

## Salary Expectations

```
salaries = [s.get('expected_salary_range_inr_lpa', {}) for s in signals]
sal_df = pd.DataFrame(salaries)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=sal_df, x='min', y='max', alpha=0.5)
plt.title('Salary Expectations (Min vs Max LPA)')
plt.plot([0, 80], [0, 80], 'r--', label='Min=Max')
plt.legend()
plt.show()
```
